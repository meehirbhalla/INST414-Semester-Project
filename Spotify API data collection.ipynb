# Credentials and access

```python
import json
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials

# credentials
cid = '39a92c9c918c4cd28997958f90a5021f'
secret = '661f653d547d430ba3d0c6b852eb6e51'

client_credentials_manager = SpotifyClientCredentials(cid, secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

# 2000/7600 part 1

```python
# playlist
url = 'https://open.spotify.com/playlist/3JIT8NyCkUVuhj4B2DdoVQ?si=baefdb24743d4324'

# user_playlist_tracks method to get JSON data of playlist
results = sp.user_playlist_tracks('meehir_b', url)

# get JSON data of tracks
tracks = results['items']

# exceed 100 track request limit
while results['next']:
    results = sp.next(results)
    tracks.extend(results['items'])
```

```python
len(tracks)
```

```python
# create a list of track ids
track_list = []

for t in tracks:
    track_list.append(t['track']['id'])
        
track_metadata = {'id':[], 'popularity':[]}

for track in track_list:

    # track metadata
    metadata = sp.track(track)

    # check for duplicates
    if track not in track_metadata['id']:
        # track unique id
        track_metadata['id'].append(track)

        # track popularity
        track_metadata['popularity'].append(metadata['popularity'])
```

```python
# check for duplicates
len(track_metadata['id']) == len(set(track_metadata['id']))
```

```python
metadata = pd.DataFrame.from_dict(track_metadata)
```

```python
# check for duplicates
metadata.duplicated()
```

audio features of each track

```python
feat = []
req = 0
x = 0
y = 100

while req < 20:
    if req == 19:
        ids = metadata['id'][x:y + 1].tolist()
    else:
        ids = metadata['id'][x:y].tolist()
    feat += (sp.audio_features(ids))
    req += 1
    x += 100
    y += 100
```

cleaning and merging of metadata and audio features

```python
# dataframe of metadata
metadata = pd.DataFrame.from_dict(track_metadata)

# dataframe of musical features
musical_features = pd.DataFrame(feat)

# merge metadata with musical features
df1 = metadata.merge(musical_features)

# drop unecessary columns
df1.drop(columns = ['track_href', 'analysis_url', 'type' ,'uri'], inplace = True)
```
